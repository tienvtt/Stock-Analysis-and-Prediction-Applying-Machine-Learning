# **Calculate Simple Moving Averages (SMA)**

# 1. What is the SMA Indicator?

SMA (**Simple Moving Average**) is a technical analysis tool used to smooth out price fluctuations and identify market trends. It calculates the average closing price of an asset over a specific period (e.g., **20 days, 50 days, 200 days**).

### Key Features:
- SMA reacts slowly to price changes because it equally weights all data points in the selected period.
- **Common timeframes:** 20, 50, 100, 200 days (**short-term, medium-term, long-term trends**).

---

# 2. Role of SMA in Stock Trading

## Trend Identification:
- **Price above SMA → Uptrend (bullish).**
- **Price below SMA → Downtrend (bearish).**

**Example:**  
The **200-day SMA** is often called the *"bull-bear line"* for long-term trends.

## Dynamic Support/Resistance:
- **In an uptrend**, SMA acts as **support** (price bounces off the SMA and rises).
- **In a downtrend**, SMA acts as **resistance** (price rallies to the SMA and drops).

## Crossover Signals:
- **Golden Cross:** Shorter SMA (e.g., 50-day) crosses above a longer SMA (e.g., 200-day) → **Buy signal**.
- **Death Cross:** Shorter SMA crosses below a longer SMA → **Sell signal**.

---

# 3. How to Analyze Stocks Using SMA

## a. Basic Price-SMA Comparison
- **Buy** when price breaks above a rising SMA.
- **Sell** when price drops below a falling SMA.

**Example:**  
If **Apple stock** stays above its 50-day SMA, traders may expect the uptrend to continue.

## b. SMA Crossovers
- **Bullish Signal:** Shorter SMA (e.g., 20-day) crosses above a longer SMA (e.g., 50-day).
- **Bearish Signal:** Shorter SMA crosses below a longer SMA.

**Example:**  
In 2020, the **S&P 500’s 50-day SMA crossing above its 200-day SMA** signaled a post-COVID recovery.

## c. Multiple SMA Strategy
Use **3 SMAs** (e.g., **20, 50, 200-day**) to confirm trends:
- **Strong uptrend:** Price > 20-day SMA > 50-day SMA > 200-day SMA.
- **Strong downtrend:** Price < 20-day SMA < 50-day SMA < 200-day SMA.

## d. Combine with Other Indicators
- **RSI:** SMA uptrend + RSI > 50 → Confirms **bullish momentum**.
- **Volume:** Rising volume during SMA crossover strengthens the signal’s reliability.

---

# 4. Limitations of SMA
- **Lagging Indicator:** SMA reacts slowly to sudden price changes, potentially delaying entry/exit points.
- **Ineffective in Sideways Markets:** SMA generates false signals when prices move horizontally.
- **Historical Data Only:** SMA reflects past trends, not future predictions. Always combine with other tools.

---

# Summary:
- **SMA simplifies trend analysis** and provides actionable signals (**crossovers, support/resistance**).
- Use SMAs of **different lengths** to confirm trends and reduce false signals.
- Always pair SMA with **volume, RSI, or MACD** for higher accuracy.


```python
def calculate_sma(df):

    # Reverse the DataFrame so the latest days are at the bottom
    df_reversed = df.sort_index(ascending=True)

    # Calculate moving averages on the reversed DataFrame
    df_reversed["SMA_10"] = df_reversed["close"].rolling(window=10).mean()
    df_reversed["SMA_20"] = df_reversed["close"].rolling(window=20).mean()
    df_reversed["SMA_50"] = df_reversed["close"].rolling(window=50).mean()

    # Reverse the DataFrame back to its original order
    # df_with_sma = df_reversed.sort_index(ascending=True)

    return df_reversed
# Apply SMA calculation
df_shb = calculate_sma(df_shb)
df_vhm = calculate_sma(df_vhm)
df_fpt= calculate_sma(df_fpt)
df_hpg = calculate_sma(df_hpg)
df_hcm = calculate_sma(df_hcm)
```

```python
def plot_sma(df, symbol, start_date):
    """Plot Closing Price & SMA from a given start date."""
    df = df[df.index >= start_date]  # Filter data from start_date

    plt.figure(figsize=(15, 6))

    plt.plot(df.index, df["close"], label="Closing Price", color="blue", alpha=0.8)
    plt.plot(df.index, df["SMA_10"], label="SMA10", color="green", alpha=0.6)
    plt.plot(df.index, df["SMA_20"], label="SMA20", color="red", alpha=0.6)
    plt.plot(df.index, df["SMA_50"], label="SMA50", color="orange", alpha=0.6)

    plt.title(f"{symbol} Closing Price & SMA Lines")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.grid()

    plt.show()

```

# **Calculate EMA**

# 1. What is the EMA Indicator?
**EMA (Exponential Moving Average)** is a technical indicator that prioritizes recent price movements. Unlike SMA, EMA assigns greater weight to the latest prices, making it more sensitive to short-term changes.

## EMA Formula:
ema.png
Where:
\[
\alpha = \frac{2}{n + 1}
\]
*(with \( n \) being the period, e.g., 20 days).*

### Example:
The **12-day and 26-day EMA** are commonly used in the MACD indicator.

### Key Features:
- **Faster reaction than SMA** due to focus on recent data.
- **Effective in volatile markets.**

---

# 2. Role of EMA in Stock Trading
## a. Short-Term Trend Identification
- EMA reacts quickly to price changes → Helps detect trend reversals early.
- **Example:** The **20-day EMA** is used for short-term trends (1-3 months).

## b. Dynamic Support/Resistance
- **In an uptrend:** EMA acts as **dynamic support** (price often bounces off EMA).
- **In a downtrend:** EMA becomes **dynamic resistance**.

## c. Crossover Signals
- **Bullish Signal:** Shorter EMA crosses above longer EMA (e.g., EMA 50 > EMA 200).
- **Bearish Signal:** Shorter EMA crosses below longer EMA.

---

# 3. SMA-Based Price Analysis
## a. Long-Term Trend Identification
- The **200-day SMA** is considered the "primary trend":
  - **Price > 200-day SMA** → Long-term **bullish**.
  - **Price < 200-day SMA** → Long-term **bearish**.

## b. SMA Crossover Strategy
- **Golden Cross:** SMA 50 crosses above SMA 200 → **Strong buy signal**.
- **Death Cross:** SMA 50 crosses below SMA 200 → **Strong sell signal**.

## c. Multiple SMA Strategy
- **SMA 20 + SMA 50 + SMA 200**:
  - **Strong uptrend:** Price > SMA 20 > SMA 50 > SMA 200.
  - **Strong downtrend:** Price < SMA 20 < SMA 50 < SMA 200.

---

# 4. EMA vs SMA Comparison

| **Criteria**         | **EMA**                          | **SMA**                         |
|----------------------|--------------------------------|--------------------------------|
| **Price Reaction**   | Fast, sensitive to recent data | Slow, stable                   |
| **Weighting**        | Prioritizes recent prices      | Equal weight to all prices     |
| **Pros**            | Early trend detection          | Fewer false signals            |
| **Cons**            | Prone to "whipsaws" (noise)    | High lag                        |


---

# Summary
**EMA:** Ideal for **short-term strategies** and quick reactions.  
**SMA:** Suitable for **long-term investors** seeking stability.  



# 1. What is the MACD Indicator?
**MACD (Moving Average Convergence Divergence)** is a momentum indicator used to identify trend direction, strength, and reversal points. It consists of three components:

- **MACD Line**: The difference between the **12-day EMA** and **26-day EMA**.
- **Signal Line**: A **9-day EMA** of the MACD Line.
- **Histogram**: The gap between the **MACD Line** and **Signal Line**.

## MACD Formula:
MACD.png

### Key Features:
- **Histogram > 0** → **Bullish momentum** (MACD Line above Signal Line).
- **Histogram < 0** → **Bearish momentum** (MACD Line below Signal Line).

---

# 2. Role of MACD in Stock Trading
## a. Trend Identification
- **MACD > 0** → **Bullish trend**.
- **MACD < 0** → **Bearish trend**.

## b. Crossover Signals
- **Buy Signal**: MACD Line **crosses above** Signal Line.
- **Sell Signal**: MACD Line **crosses below** Signal Line.

## c. Divergence
- **Bullish Divergence**: Price makes **lower lows**, but **MACD rises** → **Reversal up**.
- **Bearish Divergence**: Price makes **higher highs**, but **MACD declines** → **Reversal down**.

---

# 3. MACD-Based Price Analysis
## a. Trade MACD/Signal Line Crossovers
- **Buy:** When **MACD (blue line) crosses above Signal Line (orange line)** and Histogram turns **positive**.
- **Sell:** When **MACD crosses below Signal Line** and Histogram turns **negative**.

**Example:**  
📌 **Apple (AAPL)** surged **18%** after a bullish crossover in **March 2023**.

## b. MACD Divergence
- **Bullish Divergence**: Price forms **lower lows**, MACD forms **higher lows** → **Buy**.
- **Bearish Divergence**: Price forms **higher highs**, MACD forms **lower highs** → **Sell**.

**Example:**  
📌 **S&P 500** dropped **8%** following a **bearish divergence** in **January 2024**.

## c. Histogram as Momentum Gauge
- **Expanding Histogram** → **Strong momentum** (bullish/bearish).
- **Shrinking Histogram** → **Momentum weakening** → **Potential reversal**.

## d. Combine with RSI or SMA
- **MACD + SMA 200**:
  - **Buy** when **MACD crosses above Signal Line** and **price > SMA 200**.
  - **Sell** when **MACD crosses below Signal Line** and **price < SMA 200**.

---

# 4. Limitations of MACD
⚠ **Lagging Indicator**: Relies on EMAs, slow to react to sudden price moves.  
⚠ **False Signals**: Unreliable in sideways markets.  
⚠ **Requires Confirmation**: Always pair with **Volume, RSI, or price action**.  


# Summary
 **MACD is a versatile tool** for spotting trends, momentum, and reversals.  
 **Maximize accuracy** by combining divergence, crossovers, and Histogram analysis.  
⚠ **Avoid standalone use** in highly volatile or choppy markets.  


```python
def calculate_ema(df):
    """Calculate Exponential Moving Average (EMA)."""
    df["EMA12"] = df["close"].ewm(span=12, adjust=False).mean()
    df["EMA26"] = df["close"].ewm(span=26, adjust=False).mean()

    df["MACD"] = df["EMA12"] - df["EMA26"]
    df["MACD_Signal"] = df["MACD"].ewm(span=9, adjust=False).mean()
    return df
# Apply EMA calculation
df_shb = calculate_ema(df_shb)
df_vhm = calculate_ema(df_vhm)
df_fpt = calculate_ema(df_fpt)
df_hpg = calculate_ema(df_hpg)
df_hcm = calculate_ema(df_hcm)
```

```python
def plot_ema_macd(df, symbol, start_date):

    df = df[df.index > start_date]
    """Visualize stock trends with EMA."""
    plt.figure(figsize=(15, 6))

    plt.plot(df.index, df["close"], label=f"{symbol} Closing Price", color="blue", alpha=0.6)
    plt.plot(df.index, df["EMA12"], label="EMA12", color="orange", linestyle="dashed")
    plt.plot(df.index, df["EMA26"], label="EMA26", color="green", linestyle="dashed")

    plt.title(f"{symbol} Stock Price & EMA Line", fontsize=14)
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.grid()

    plt.show()
```

# **Calculate RSI**

# 1. What is the RSI Indicator?
RSI (Relative Strength Index) is a momentum oscillator that measures the speed and magnitude of price movements, identifying overbought or oversold conditions. It ranges from 0 to 100, typically using a 14-day period.

RSI.png

<br>**Example:**  
If prices rise for 7 days and fall for 7 days, the RSI reflects a balance between buying and selling pressure.

## Key Features:
- **Overbought:** RSI > 70 → Potential price correction.
- **Oversold:** RSI < 30 → Potential price rebound.
- **Divergence:** Signals potential trend reversal.

---

# 2. Role of RSI in Stock Trading

## a. Identify Overbought/Oversold Conditions
- **RSI > 70:** Market may correct due to excessive buying.
- **RSI < 30:** Market may rebound due to excessive selling.

## b. Detect Divergence
- **Bullish Divergence:** Price makes lower lows, but RSI makes higher lows → **Buy signal**.
- **Bearish Divergence:** Price makes higher highs, but RSI makes lower highs → **Sell signal**.

## c. Confirm Trend Strength
- **RSI > 50:** Bullish trend dominates.
- **RSI < 50:** Bearish trend dominates.

---

# 3. RSI-Based Price Analysis

## a. Trade Overbought/Oversold Zones
- **Buy:** When RSI crosses above 30 (exits oversold zone).
- **Sell:** When RSI crosses below 70 (exits overbought zone).  

**Example:**  
Apple (AAPL) rallied **10%** after RSI hit **25 (oversold)** in April 2023.

## b. RSI Divergence
- **Buy Signal:** Bullish divergence during a downtrend.
- **Sell Signal:** Bearish divergence during an uptrend.

**Example:**  
S&P 500 made a new high in January 2024, but RSI declined → **8% drop followed**.

## c. RSI Trendline Breakouts
- **RSI breaks a downtrend line** → **Bullish signal**.
- **RSI breaks an uptrend line** → **Bearish signal**.

## d. Combine with SMA/EMA
**RSI + SMA 50:**
- **Buy:** When RSI > 50 and price crosses above SMA 50.
- **Sell:** When RSI < 50 and price crosses below SMA 50.

---

# 4. Limitations of RSI
- **False Signals (Whipsaws):** RSI can stay in overbought/oversold zones during strong trends.
- **Ineffective in Sideways Markets:** RSI fluctuates between 30–70 without clear signals.
- **Requires Confirmation:** Pair with MACD, Volume, or Bollinger Bands for reliability.

---


# Summary
- RSI is a powerful momentum tool for spotting **overbought/oversold conditions** and **divergence**.
- Combine RSI with **trendlines, SMA, or volume** to refine signals.
- Avoid relying solely on RSI in **highly volatile markets**.


```python
def calculate_rsi(df, period=14):
    """Calculate Relative Strength Index (RSI)."""
    delta = df["close"].diff(1)
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)

    avg_gain = gain.rolling(window=period, min_periods=1).mean()
    avg_loss = loss.rolling(window=period, min_periods=1).mean()

    rs = avg_gain / (avg_loss + 1e-10)  # Prevent division by zero
    df["RSI"] = 100 - (100 / (1 + rs))
    return df

# Apply RSI calculation
df_shb = calculate_rsi(df_shb)
df_vhm = calculate_rsi(df_vhm)
df_fpt = calculate_rsi(df_fpt)
df_hpg = calculate_rsi(df_hpg)
df_hcm = calculate_rsi(df_hcm)
```

```python
def plot_macd_rsi(df, symbol, start_date):

    df = df[df.index >= start_date]
    """Visualize MACD and RSI indicators."""
    fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 8), sharex=True)

    # MACD Plot
    axes[0].plot(df.index, df["MACD"], label="MACD", color="blue")
    axes[0].plot(df.index, df["MACD_Signal"], label="Signal Line", color="red", linestyle="dashed")
    axes[0].bar(df.index, df["MACD"] - df["MACD_Signal"], color=(df["MACD"] > df["MACD_Signal"]).map({True: "green", False: "red"}), alpha=0.5)
    axes[0].set_title(f"{symbol} - MACD Indicator")
    axes[0].legend()
    axes[0].grid()

    # RSI Plot
    axes[1].plot(df.index, df["RSI"], label="RSI", color="purple")
    axes[1].axhline(70, linestyle="dashed", color="red", alpha=0.5)  # Overbought level
    axes[1].axhline(30, linestyle="dashed", color="green", alpha=0.5)  # Oversold level
    axes[1].set_title(f"{symbol} - RSI Indicator")
    axes[1].legend()
    axes[1].grid()

    plt.tight_layout()
    plt.show()
```
